## 📚 Теоретическая справка

### Линейная регрессия
Модель предсказывает целевую переменную как взвешенную сумму признаков:

$$y = w_1x_1 + w_2x_2 + ... + w_nx_n + b$$

### Функция потерь MSE
$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

### Градиентный спуск
$$w_{new} = w_{old} - \eta \cdot \nabla MSE$$

где $\eta$ — скорость обучения (learning rate)

### Задание 2: Реализация градиентного спуска с нуля

````python
class GradientDescentLinearRegression:
    """
    Реализация линейной регрессии с градиентным спуском с нуля
    """

    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None
        self.loss_history = []

    def fit(self, X, y):
        """
        Обучение модели с помощью градиентного спуска
        """
        n_samples, n_features = X.shape

        # Инициализация параметров
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Градиентный спуск
        for i in range(self.n_iterations):
            # Предсказание
            y_pred = np.dot(X, self.weights) + self.bias

            # Вычисление градиентов
            dw = (2/n_samples) * np.dot(X.T, (y_pred - y))
            db = (2/n_samples) * np.sum(y_pred - y)

            # Обновление параметров
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

            # Сохранение истории потерь
            mse = np.mean((y_pred - y) ** 2)
            self.loss_history.append(mse)

            # Вывод прогресса каждые 100 итераций
            if i % 100 == 0:
                print(f"Итерация {i}: MSE = {mse:.4f}")

        return self

    def predict(self, X):
        """
        Предсказание целевой переменной
        """
        return np.dot(X, self.weights) + self.bias

    def score(self, X, y):
        """
        Вычисление R² score
        """
        y_pred = self.predict(X)
        ss_res = np.sum((y - y_pred) ** 2)
        ss_tot = np.sum((y - np.mean(y)) ** 2)
        return 1 - (ss_res / ss_tot)
````

````python
# Обучаем нашу модель
print("Обучение модели с градиентным спуском...")
start_time = time.time()

gd_model = GradientDescentLinearRegression(learning_rate=0.01, n_iterations=1000)
gd_model.fit(X_train_scaled, y_train)

gd_time = time.time() - start_time
print(f"\nВремя обучения: {gd_time:.4f} секунд")
print(f"Веса модели: {gd_model.weights.round(4)}")
print(f"Смещение (bias): {gd_model.bias:.4f}")
````

````python
# Интерактивная визуализация влияния learning rate
from ipywidgets import interact, FloatSlider, IntSlider

def plot_gradient_descent_interactive(learning_rate=0.01, n_iterations=100):
    """
    Интерактивная демонстрация влияния параметров на сходимость
    """
    model = GradientDescentLinearRegression(
        learning_rate=learning_rate,
        n_iterations=n_iterations
    )
    model.fit(X_train_scaled[:100], y_train[:100])  # Используем подвыборку для скорости

    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(model.loss_history, linewidth=2, color='blue')
    plt.xlabel('Итерация', fontsize=12)
    plt.ylabel('MSE', fontsize=12)
    plt.title(f'LR={learning_rate}, Iter={n_iterations}', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.yscale('log')

    plt.subplot(1, 2, 2)
    plt.bar(range(len(model.weights)), model.weights, color='green', alpha=0.7)
    plt.xlabel('Признак', fontsize=12)
    plt.ylabel('Вес', fontsize=12)
    plt.title('Веса модели', fontsize=14)
    plt.grid(True, alpha=0.3, axis='y')

    plt.tight_layout()
    plt.show()

    print(f"Финальное MSE: {model.loss_history[-1]:.4f}")
    print(f"R² score: {model.score(X_train_scaled[:100], y_train[:100]):.4f}")

# Создаём интерактивный виджет
interact(plot_gradient_descent_interactive,
         learning_rate=FloatSlider(value=0.01, min=0.001, max=0.1, step=0.001),
         n_iterations=IntSlider(value=100, min=10, max=500, step=10));
````

### Задание 4: Сравнение трёх подходов

````python
# 1. Градиентный спуск (наша реализация)
start_time = time.time()
gd_model = GradientDescentLinearRegression(learning_rate=0.01, n_iterations=1000)
gd_model.fit(X_train_scaled, y_train)
gd_pred = gd_model.predict(X_test_scaled)
gd_time = time.time() - start_time

# 2. Нормальное уравнение (аналитическое решение)
start_time = time.time()
X_train_bias = np.column_stack([np.ones(X_train_scaled.shape[0]), X_train_scaled])
# w = (X^T X)^(-1) X^T y
normal_weights = np.linalg.inv(X_train_bias.T.dot(X_train_bias)).dot(X_train_bias.T).dot(y_train)
normal_pred = np.column_stack([np.ones(X_test_scaled.shape[0]), X_test_scaled]).dot(normal_weights)
normal_time = time.time() - start_time

# 3. LinearRegression из sklearn
start_time = time.time()
sklearn_model = LinearRegression()
sklearn_model.fit(X_train_scaled, y_train)
sklearn_pred = sklearn_model.predict(X_test_scaled)
sklearn_time = time.time() - start_time

# Сравниваем результаты
print("=" * 70)
print("СРАВНЕНИЕ ТРЁХ ПОДХОДОВ")
print("=" * 70)

methods = ['Градиентный спуск', 'Нормальное уравнение', 'sklearn LinearRegression']
times = [gd_time, normal_time, sklearn_time]
mse_scores = [
    mean_squared_error(y_test, gd_pred),
    mean_squared_error(y_test, normal_pred),
    mean_squared_error(y_test, sklearn_pred)
]
r2_scores = [
    r2_score(y_test, gd_pred),
    r2_score(y_test, normal_pred),
    r2_score(y_test, sklearn_pred)
]

comparison_df = pd.DataFrame({
    'Метод': methods,
    'Время (сек)': [f'{t:.4f}' for t in times],
    'MSE': [f'{m:.4f}' for m in mse_scores],
    'R²': [f'{r:.4f}' for r in r2_scores]
})

print(comparison_df.to_string(index=False))
print("=" * 70)
````

##  Технические требования к генерации данных

```python
def generate_dataset(n_samples=500, n_features=2, noise_std=0.5,
                     outlier_fraction=0.05, outlier_multiplier=3,
                     random_seed=None):
    """
    Генерация синтетического датасета для линейной регрессии
    
    Parameters:
    -----------
    n_samples : int - количество наблюдений
    n_features : int - количество признаков
    noise_std : float - стандартное отклонение шума
    outlier_fraction : float - доля выбросов (0.0-0.2)
    outlier_multiplier : float - множитель для выбросов
    random_seed : int - seed для воспроизводимости
    """
    if random_seed is not None:
        np.random.seed(random_seed)
    
    # Генерация признаков
    X = np.random.randn(n_samples, n_features)
    
    # Истинные коэффициенты
    true_coef = np.random.uniform(1, 5, n_features)
    true_intercept = np.random.uniform(0, 10)
    
    # Генерация целевой переменной
    y = X @ true_coef + true_intercept + np.random.randn(n_samples) * noise_std
    
    # Добавление выбросов
    n_outliers = int(n_samples * outlier_fraction)
    outlier_idx = np.random.choice(n_samples, n_outliers, replace=False)
    y[outlier_idx] += np.random.choice([-1, 1], n_outliers) * \
                      np.random.uniform(3, 10, n_outliers) * outlier_multiplier
    
    return X, y, true_coef, true_intercept
```

    for name, model in models_dict.items():

            'MSE': mean_squared_error(y_test, y_pred),
            'MAE': mean_absolute_error(y_test, y_pred),
            'R2': r2_score(y_test, y_pred)
        })
    return pd.DataFrame(results)

````python
# -*- coding: utf-8 -*-
"""Практическая_работа_3_Вариант_9.ipynb

Automatically generated by Colaboratory.

## Практическая работа №3: Линейная регрессия и функции потерь (Вариант 9)
"""

# Commented out IPython magic to ensure Python compatibility.
# %%capture
# Импортируем необходимые библиотеки
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, HuberRegressor, SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

# Настройка стиля графиков
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 5)

# ----------------------------------------------------------------------------
# Этап 1: Генерация и подготовка данных
# ----------------------------------------------------------------------------

print("ЭТАП 1: ГЕНЕРАЦИЯ И ПОДГОТОВКА ДАННЫХ")

# Параметры из варианта 9
student_id = 9
n_samples = 500
n_features = 2
feature_names = ['seo_score', 'content_len']
target_name = 'website_traffic'
noise_std = 0.4
outlier_fraction = 0.04
outlier_multiplier = 2.5
epsilon_huber = 2.5
loss_sgd = 'squared_loss'
learning_rate_sgd = 0.01
n_iter_sgd = 1100

# Устанавливаем seed для воспроизводимости
np.random.seed(student_id)

def generate_dataset(n_samples=500, n_features=2, noise_std=0.5,
                     outlier_fraction=0.05, outlier_multiplier=3,
                     random_seed=None):
    """
    Генерация синтетического датасета для линейной регрессии
    """
    if random_seed is not None:
        np.random.seed(random_seed)

    # Генерация признаков из нормального распределения
    X = np.random.randn(n_samples, n_features)

    # Истинные коэффициенты и свободный член
    true_coef = np.random.uniform(1, 5, n_features)
    true_intercept = np.random.uniform(0, 10)

    # Генерация целевой переменной (линейная комбинация + шум)
    y = X @ true_coef + true_intercept + np.random.randn(n_samples) * noise_std

    # Добавление выбросов
    n_outliers = int(n_samples * outlier_fraction)
    outlier_idx = np.random.choice(n_samples, n_outliers, replace=False)
    # Добавляем выбросы с заданной амплитудой и случайным знаком
    y[outlier_idx] += np.random.choice([-1, 1], n_outliers) * \
                      np.random.uniform(3, 10, n_outliers) * outlier_multiplier

    return X, y, true_coef, true_intercept

# Генерация данных
X, y, true_coef, true_intercept = generate_dataset(
    n_samples=n_samples,
    n_features=n_features,
    noise_std=noise_std,
    outlier_fraction=outlier_fraction,
    outlier_multiplier=outlier_multiplier,
    random_seed=student_id
)

print(f"Размер датасета: {X.shape}")
print(f"Количество признаков: {X.shape[1]}")
print(f"Признаки: {feature_names}")
print(f"Целевая переменная: {target_name}")
print(f"\nИстинные коэффициенты (генерируемые):")
for name, coef in zip(feature_names, true_coef):
    print(f"  {name}: {coef:.4f}")
print(f"  intercept: {true_intercept:.4f}")

# Создаём DataFrame для удобства анализа
df = pd.DataFrame(X, columns=feature_names)
df[target_name] = y

print(f"\nПервые 5 строк сгенерированных данных:")
print(df.head())

# Разделение на обучающую и тестовую выборки (70/30)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f"\nОбучающая выборка: {X_train.shape}")
print(f"Тестовая выборка: {X_test.shape}")

# Масштабирование признаков (критически важно для SGD!)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nМасштабирование выполнено!")
print(f"Среднее после масштабирования (train): {np.mean(X_train_scaled, axis=0).round(2)}")
print(f"Ст. отклонение после масштабирования (train): {np.std(X_train_scaled, axis=0).round(2)}")

# ----------------------------------------------------------------------------
# Этап 2: Разведочный анализ данных (EDA)
# ----------------------------------------------------------------------------

print("\nЭТАП 2: РАЗВЕДОЧНЫЙ АНАЛИЗ ДАННЫХ (EDA)")

# 1. Матрица корреляций
plt.figure(figsize=(8, 6))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm',
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Матрица корреляций признаков и целевой переменной', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# 2. Pairplot для визуализации зависимостей
print("\nPairplot для визуализации попарных зависимостей:")
sns.pairplot(df)
plt.suptitle('Попарные зависимости переменных', y=1.02, fontsize=14, fontweight='bold')
plt.show()

# 3. Анализ распределения целевой переменной
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df[target_name], kde=True, color='skyblue', bins=30)
plt.xlabel(target_name, fontsize=12)
plt.ylabel('Частота', fontsize=12)
plt.title('Распределение целевой переменной', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
sns.boxplot(y=df[target_name], color='lightcoral')
plt.ylabel(target_name, fontsize=12)
plt.title('Box-plot целевой переменной (поиск выбросов)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

# 4. Визуализация выбросов (для первых двух признаков)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for i, feature in enumerate(feature_names):
    axes[i].scatter(df[feature], df[target_name], alpha=0.6, label='Нормальные данные')
    axes[i].set_xlabel(feature, fontsize=12)
    axes[i].set_ylabel(target_name, fontsize=12)
    axes[i].set_title(f'{feature} vs {target_name}', fontsize=14, fontweight='bold')
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ----------------------------------------------------------------------------
# Этап 3: Обучение моделей
# ----------------------------------------------------------------------------

print("\nЭТАП 3: ОБУЧЕНИЕ МОДЕЛЕЙ")

# 1. LinearRegression (обычная MSE)
print("\n--- Обучение LinearRegression (MSE) ---")
model_lr = LinearRegression()
model_lr.fit(X_train_scaled, y_train)
print(f"Коэффициенты: {model_lr.coef_.round(4)}")
print(f"Свободный член (intercept): {model_lr.intercept_:.4f}")

# 2. HuberRegressor (робастная регрессия)
print(f"\n--- Обучение HuberRegressor (epsilon={epsilon_huber}) ---")
model_huber = HuberRegressor(epsilon=epsilon_huber, max_iter=1000)
model_huber.fit(X_train_scaled, y_train)
print(f"Коэффициенты: {model_huber.coef_.round(4)}")
print(f"Свободный член (intercept): {model_huber.intercept_:.4f}")

# 3. SGDRegressor (градиентный спуск)
print(f"\n--- Обучение SGDRegressor (loss='{loss_sgd}', learning_rate={learning_rate_sgd}) ---")
model_sgd = SGDRegressor(loss=loss_sgd, max_iter=n_iter_sgd, eta0=learning_rate_sgd,
                         learning_rate='constant', random_state=42)
model_sgd.fit(X_train_scaled, y_train)
print(f"Коэффициенты: {model_sgd.coef_.round(4)}")
print(f"Свободный член (intercept): {model_sgd.intercept_:.4f}")

# ----------------------------------------------------------------------------
# Этап 4: Оценка качества
# ----------------------------------------------------------------------------

print("\nЭТАП 4: ОЦЕНКА КАЧЕСТВА МОДЕЛЕЙ")

# Предсказания
y_pred_lr = model_lr.predict(X_test_scaled)
y_pred_huber = model_huber.predict(X_test_scaled)
y_pred_sgd = model_sgd.predict(X_test_scaled)

# Сбор метрик
models = ['LinearRegression', 'HuberRegressor', 'SGDRegressor']
predictions = [y_pred_lr, y_pred_huber, y_pred_sgd]

results = []
for name, y_pred in zip(models, predictions):
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results.append({
        'Модель': name,
        'MSE': mse,
        'RMSE': rmse,
        'MAE': mae,
        'R²': r2
    })

results_df = pd.DataFrame(results)
print("\nСводная таблица метрик качества:")
print(results_df.to_string(index=False))

# Визуализация сравнения метрик
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

metrics_to_plot = ['MSE', 'MAE', 'R²']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

for i, metric in enumerate(metrics_to_plot):
    axes[i].bar(results_df['Модель'], results_df[metric], color=colors, alpha=0.7)
    axes[i].set_xlabel('Модель', fontsize=12)
    axes[i].set_ylabel(metric, fontsize=12)
    axes[i].set_title(f'Сравнение моделей по метрике {metric}', fontsize=14, fontweight='bold')
    axes[i].grid(True, alpha=0.3, axis='y')
    # Добавляем значения на столбцы
    for j, v in enumerate(results_df[metric]):
        axes[i].text(j, v + 0.02, f'{v:.4f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# График "Фактические vs Предсказанные значения" для всех моделей
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, (name, y_pred) in enumerate(zip(models, predictions)):
    axes[i].scatter(y_test, y_pred, alpha=0.5, edgecolors='black', linewidth=0.5)
    # Линия идеального предсказания
    min_val = min(y_test.min(), y_pred.min())
    max_val = max(y_test.max(), y_pred.max())
    axes[i].plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='Идеальное предсказание')
    axes[i].set_xlabel('Реальные значения', fontsize=12)
    axes[i].set_ylabel('Предсказанные значения', fontsize=12)
    axes[i].set_title(f'{name}', fontsize=14, fontweight='bold')
    axes[i].legend(loc='upper left')
    axes[i].grid(True, alpha=0.3)
    axes[i].set_aspect('equal', adjustable='box')
    # Добавляем метрику R² на график
    axes[i].text(0.05, 0.95, f'R² = {results_df.loc[results_df["Модель"]==name, "R²"].values[0]:.3f}',
                 transform=axes[i].transAxes, fontsize=11, verticalalignment='top',
                 bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.suptitle('Фактические vs Предсказанные значения', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# График остатков для всех моделей
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, (name, y_pred) in enumerate(zip(models, predictions)):
    residuals = y_test - y_pred
    axes[i].scatter(y_pred, residuals, alpha=0.5, edgecolors='black', linewidth=0.5)
    axes[i].axhline(y=0, color='red', linestyle='--', linewidth=2, label='Нулевая линия')
    axes[i].set_xlabel('Предсказанные значения', fontsize=12)
    axes[i].set_ylabel('Остатки (y_test - y_pred)', fontsize=12)
    axes[i].set_title(f'Остатки модели ({name})', fontsize=14, fontweight='bold')
    axes[i].legend(loc='upper left')
    axes[i].grid(True, alpha=0.3)
    # Добавляем метрику MAE на график
    axes[i].text(0.05, 0.95, f'MAE = {results_df.loc[results_df["Модель"]==name, "MAE"].values[0]:.3f}',
                 transform=axes[i].transAxes, fontsize=11, verticalalignment='top',
                 bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.5))

plt.suptitle('Анализ остатков', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# ----------------------------------------------------------------------------
# Этап 5: Анализ устойчивости
# ----------------------------------------------------------------------------

print("\nЭТАП 5: АНАЛИЗ УСТОЙЧИВОСТИ")

# 1. Исследование влияния epsilon на HuberRegressor
print("\n--- Исследование влияния параметра epsilon на HuberRegressor ---")
epsilon_values = [1.0, 1.35, 1.5, 2.0, 2.5, 3.0, 4.0, 5.0]
huber_results = []

for eps in epsilon_values:
    model = HuberRegressor(epsilon=eps, max_iter=1000)
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    huber_results.append({
        'epsilon': eps,
        'MSE': mean_squared_error(y_test, y_pred),
        'MAE': mean_absolute_error(y_test, y_pred),
        'R2': r2_score(y_test, y_pred)
    })

huber_results_df = pd.DataFrame(huber_results)
print("\nВлияние epsilon на метрики качества:")
print(huber_results_df.to_string(index=False))

# Визуализация влияния epsilon
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
metrics = ['MSE', 'MAE', 'R2']
for i, metric in enumerate(metrics):
    axes[i].plot(huber_results_df['epsilon'], huber_results_df[metric], marker='o', linestyle='-', color='teal')
    axes[i].set_xlabel('epsilon', fontsize=12)
    axes[i].set_ylabel(metric, fontsize=12)
    axes[i].set_title(f'Влияние epsilon на {metric}', fontsize=14, fontweight='bold')
    axes[i].grid(True, alpha=0.3)
    # Отмечаем значение из варианта
    axes[i].axvline(x=epsilon_huber, color='red', linestyle='--', linewidth=1.5, label=f'Ваш epsilon ({epsilon_huber})')
    axes[i].legend()

plt.suptitle('Анализ устойчивости HuberRegressor к параметру epsilon', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# 2. Сравнение коэффициентов моделей с истинными значениями
print("\n--- Сравнение коэффициентов моделей с истинными значениями ---")
coef_comparison = pd.DataFrame({
    'Признак': feature_names,
    'Истинный': true_coef,
    'LinearRegression': model_lr.coef_,
    'HuberRegressor': model_huber.coef_,
    'SGDRegressor': model_sgd.coef_
})
print(coef_comparison.to_string(index=False))
print(f"\nИстинный intercept: {true_intercept:.4f}")
print(f"LinearRegression intercept: {model_lr.intercept_:.4f}")
print(f"HuberRegressor intercept: {model_huber.intercept_:.4f}")
print(f"SGDRegressor intercept: {model_sgd.intercept_:.4f}")

# Визуализация сравнения коэффициентов
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(feature_names))
width = 0.2

ax.bar(x - 1.5*width, true_coef, width, label='Истинные', color='green', alpha=0.7)
ax.bar(x - 0.5*width, model_lr.coef_, width, label='LinearRegression', color='blue', alpha=0.7)
ax.bar(x + 0.5*width, model_huber.coef_, width, label='HuberRegressor', color='orange', alpha=0.7)
ax.bar(x + 1.5*width, model_sgd.coef_, width, label='SGDRegressor', color='purple', alpha=0.7)

ax.set_xlabel('Признаки', fontsize=12)
ax.set_ylabel('Значение коэффициента', fontsize=12)
ax.set_title('Сравнение коэффициентов моделей с истинными значениями', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(feature_names)
ax.legend()
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

# ----------------------------------------------------------------------------
# Этап 6: Выводы
# ----------------------------------------------------------------------------

print("\nЭТАП 6: ВЫВОДЫ")

print("""
На основе проведенного анализа можно сделать следующие выводы:

1.  **Влияние выбросов на модели:**
    *   Обычная линейная регрессия (LinearRegression), минимизирующая MSE,
        оказалась наиболее чувствительной к выбросам. Это видно по тому,
        как её коэффициенты отклонились от истинных значений, а также по
        сравнительно высокому MSE на тестовой выборке.
    *   HuberRegressor, благодаря комбинации MSE и MAE в функции потерь,
        показал себя значительно более устойчивым к выбросам. Его коэффициенты
        ближе к истинным, а метрики качества на тесте лучше.
    *   SGDRegressor с loss='squared_loss' показал себя так же, как и
        `LinearRegression`, поскольку оба оптимизируют одну и ту же функцию потерь,
        но с использованием разных методов (аналитическое решение против
        градиентного спуска).

2.  **Влияние параметра epsilon в HuberRegressor:**
    *   Анализ показал, что при малых значениях `epsilon` (близких к 1.0)
        модель ведет себя подобно MAE, штрафуя большие ошибки линейно, что
        делает её очень робастной.
    *   При увеличении `epsilon` (например, до 5.0 и выше) поведение Huber
        приближается к MSE, так как все большая часть ошибок попадает в
        квадратичную область. Модель становится менее робастной, но точнее
        на данных без выбросов.
    *   Для данного набора данных значение `epsilon = {epsilon_huber}` (из
        варианта) является хорошим компромиссом, обеспечивающим устойчивость
        к выбросам и достаточно высокую точность.

3.  **Сравнение функций потерь:**
    *   **MSE (квадратичная):** Проста для оптимизации, но очень чувствительна
        к выбросам. Подходит для данных, где выбросы отсутствуют или были
        предварительно обработаны.
    *   **Huber (комбинированная):** Оптимальный выбор для данных с выбросами.
        Позволяет сохранить математические свойства MSE для малых ошибок и
        робастность MAE для больших. Параметр `epsilon` позволяет гибко
        настраивать модель.
    *   **SGD с `squared_loss`:** По сути, это та же MSE, но реализованная
        через стохастический градиентный спуск. Её преимущество - возможность
        масштабирования на очень большие данные, которые не помещаются в память.

4.  **Применимость моделей:**
    *   Для задачи прогнозирования веб-трафика (`website_traffic`), где выбросы
        (например, всплески трафика в дни рекламных акций) вполне вероятны,
        наилучшей моделью является **HuberRegressor**. Она сочетает в себе
        хорошую точность и устойчивость к аномалиям.
    *   `LinearRegression` можно использовать только после тщательной
        предварительной обработки и удаления выбросов.
    *   `SGDRegressor` будет незаменим при работе с потоками данных или
        датасетами, размер которых превышает объем оперативной памяти.

5.  **Масштабирование признаков:**
    *   Эксперимент подтвердил критическую важность масштабирования для
        `SGDRegressor`, так как без него градиенты по разным признакам
        имели бы разный масштаб, что замедлило бы или вовсе предотвратило
        сходимость алгоритма.
""".format(epsilon_huber=epsilon_huber))
````
